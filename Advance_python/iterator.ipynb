An iterator is an object that allows you to go through elements of a collection one at a time.

Think of it like reading a book page by page instead of loading the entire book into your hands at once.


```python
my_list=[1,2,3,4,5,]
for i in my_list:    print(i)
```

```python
print(type(my_list))
print(type(iter(my_list)))
```

```python
print(my_list)
```

```python
it=iter(my_list)
print(it)
```

```python
next(it)
```

```python
try:
    while True:
        print(next(it))
except StopIteration:
    print("End of iterator reached.")        
```

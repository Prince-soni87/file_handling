##Pandas DataFrame And Series

Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

In Pandas, **`loc`** and **`iloc`** are both used to select rows and columns, but they differ in how they identify data.

| Feature        | `loc`                             | `iloc`                             |
| -------------- | --------------------------------- | ---------------------------------- |
| Selection type | Label-based                       | Integer position-based             |
| Uses           | Row/column names (labels)         | Row/column indexes (0,1,2,...)     |
| End index      | Inclusive                         | Exclusive                          |
| Syntax         | `df.loc[row_label, column_label]` | `df.iloc[row_index, column_index]` |

### Example DataFrame

```python
import pandas as pd

data = {
    'Name': ['Krish', 'John', 'Jack'],
    'Age': [25, 30, 45],
    'City': ['Bangalore', 'New York', 'Florida']
}

df = pd.DataFrame(data)

print(df)
```

Output:

```text
    Name  Age       City
0  Krish   25  Bangalore
1   John   30   New York
2   Jack   45    Florida
```

---

## Using `loc`

Select by **label**:

```python
print(df.loc[0])
```

Output:

```text
Name         Krish
Age             25
City     Bangalore
```

Select specific row and column:

```python
print(df.loc[1, 'Name'])
```

Output:

```text
John
```

Multiple rows and columns:

```python
print(df.loc[0:1, ['Name', 'Age']])
```

Output:

```text
    Name  Age
0  Krish   25
1   John   30
```

Notice that `0:1` includes both rows **0 and 1**.

---

## Using `iloc`

Select by **position**:

```python
print(df.iloc[0])
```

Output:

```text
Name         Krish
Age             25
City     Bangalore
```

Select specific row and column:

```python
print(df.iloc[1, 0])
```

Output:

```text
John
```

Multiple rows and columns:

```python
print(df.iloc[0:2, 0:2])
```

Output:

```text
    Name  Age
0  Krish   25
1   John   30
```

Notice that `0:2` includes rows **0 and 1** but excludes row **2**.

---

## When Custom Indexes Exist

```python
df.index = ['a', 'b', 'c']
print(df)
```

Output:

```text
    Name  Age       City
a  Krish   25  Bangalore
b   John   30   New York
c   Jack   45    Florida
```

### `loc` uses labels

```python
df.loc['b']
```

Returns row with label `'b'`.

### `iloc` uses positions

```python
df.iloc[1]
```

Returns the second row (John).

---

### Quick Rule

* Use **`loc`** when you know the **row/column names**.
* Use **`iloc`** when you know the **row/column positions**.

A common interview question:

```python
df.loc[0:2]   # returns rows 0,1,2
df.iloc[0:2]  # returns rows 0,1
```

The key difference is that **`loc` includes the ending index, while `iloc` excludes it**.

````python
### Merging and Joining DataFrames

# Create sample DataFrames

df1 = pd.DataFrame({
    'Key': ['A', 'B', 'C'],
    'Value1': [1, 2, 3]
})

df2 = pd.DataFrame({
    'Key': ['A', 'B', 'D'],
    'Value2': [4, 5, 6]
})
df1
````

````python
df2
````

````python
merged_df = pd.merge(df1, df2, on='Key',how='inner')

print(merged_df)
````

````python
merged_df = pd.merge(df1, df2, on='Key',how='outer')

print(merged_df)
````

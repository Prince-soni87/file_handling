In a pie chart, explode is used to move one or more slices away from the center to highlight them.

Your code
labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 40, 10]

explode = (0.2, 0, 0, 0)

The values in explode correspond to the slices A, B, C, D.

0.2 → Slice A is moved outward by 20% of the pie radius.
0 → Slice B stays in its normal position.
0 → Slice C stays in its normal position.
0 → Slice D stays in its normal position.

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 40, 50]
colors = ['gold', 'yellow', 'red', 'blue']
explode = (0.2, 0, 0, 0)## move out first slice from radius 

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    shadow=True
)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
tips = sns.load_dataset('tips')

# Scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips)

plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
sns.lineplot(x='size',y='total_bill',data=tips)
```

```python
tips
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar plot")
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)## kde for smothness of these data 
```

```python
sns.kdeplot(tips['total_bill'],shade=True)
```

```python
sns.pairplot(tips)
```

```python
## HeatMap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

# Interview Questions

### Q1. What is a Process?

**Answer:**

A process is an instance of a program that is currently being executed by the operating system.

---

### Q2. Difference Between Program and Process?

**Answer:**

A program is a static file stored on disk, whereas a process is a running instance of that program in memory.

---

### Q3. What is PID?

**Answer:**

PID (Process ID) is a unique identifier assigned by the operating system to each running process.

---

### Q4. Why Does Each Process Have Separate Memory?

**Answer:**

Separate memory prevents one process from accessing or corrupting another process's data, improving security and stability.

---

### Q5. What is Context Switching?

**Answer:**

Context switching is the process of saving the state of one process and loading the state of another so the CPU can execute multiple processes seemingly simultaneously.

# Memory Sharing

## Process Memory

Each process gets its own:

```text
Code Segment
Data Segment
Heap
Stack
Registers
```

---

## In Multi-Threading

Threads SHARE:

```text
✓ Code Segment
✓ Data Segment
✓ Heap
```

Each thread has its OWN:

```text
✓ Stack
✓ Registers
```

### Why Separate Stack?

Suppose:

```python
def add():
    a = 10
```

Thread 1:

```text
a = 10
```

Thread 2:

```text
a = 20
```

If they shared the same stack:

```text
Data Corruption
```

would occur.

Therefore every thread gets:

```text
Own Stack
Own Registers
```

# Interview Questions

### Q1. What is a Thread?

**Answer:**

A thread is the smallest unit of execution within a process. It performs a specific task while sharing the process's resources.

---

### Q2. Difference Between Process and Thread?

| Process                       | Thread                           |
| ----------------------------- | -------------------------------- |
| Independent program execution | Unit of execution inside process |
| Own memory space              | Shares process memory            |
| Heavyweight                   | Lightweight                      |
| Slower creation               | Faster creation                  |
| Has PID                       | Uses process resources           |

---

### Q3. What Memory is Shared Between Threads?

**Answer:**

Threads share:

```text
Code Segment
Data Segment
Heap
```

Each thread has its own:

```text
Stack
Registers
```

---

### Q4. Why are Threads Faster?

**Answer:**

Threads share the same process resources and memory, so creating and switching between threads requires less overhead than processes.

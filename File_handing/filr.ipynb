```python
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ## strip() is used to remove any leading or trailing whitespace characters from the line before printing it.
```

```python
## Write a file with (Overwriting)
with open('example.txt','w') as file:
    file.write('Hello World \n')
    file.write('this is new file\n')
    
```

```python
## Write a file without overwriting 
with open('example.txt','a') as file:
    file.write('This line will be added to the existing content.\n')
```

```python
## Write a list of line to a file
Lines=['these is my firsr line\n', 'this is my second Line\n','this is my third line\n']
with open('example.txt','a') as file:
    file.writelines(Lines)
```

```python
##Read a file and store its content in a list
with open('example.txt','r') as file:
    content=file.read()
    lines=file.readlines()
    print(lines)

with open('destination.txt','w') as file:
    file.write(content)    
```

```python
def read_file(file_path):
    with open(file_path,'r') as file:
        lines=file.readlines()
        lines_count=len(lines)
        line_word=sum( len(line.split())for line in lines)
        line_character=sum(len(line) for line in lines)
        return lines_count,line_word,line_character
file_path='example.txt'
lines_count,line_word,line_character=read_file(file_path)
print(f"number of lines:{lines_count}")
print(f"number of words:{line_word}")
print(f"number of characters:{line_character}")
```

w+ mode in python  is used to open a file  for both reading and writing . if  the file does not exit , it will be created .if file is exit then content is truncated(content of file is overwrite)


```python
with open('example.txt','w+') as file:
    file.write('hello world\n')
    file.write('this is new file\n')
    file.seek(0)
    content=file.read()
    print(content)
```

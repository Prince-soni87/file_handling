SQL(Structured Query Language) is a standard language for managing and manipulating relational database.

SQLite is a self-contained,serverless  and zero-configuration database engine that is widely used for embedded database system.


SQLite is a lightweight database that stores data in a single file.

Unlike MySQL or PostgreSQL, you don't need to install a separate database server.

Python provides a built-in module:

import sqlite3
2. What is a Connection?

A Connection is like opening the door to the database.

import sqlite3

conn = sqlite3.connect("student.db")
What happens?
If student.db exists → connect to it.
If it doesn't exist → create it.

Think of it as:

Database File
      ↑
   Connection
      ↑
   Python Program

Without a connection, you cannot read or write data.

3. What is a Cursor?

A Cursor is used to execute SQL queries.

cursor = conn.cursor()

Think of it as a messenger between Python and the database.

Python
   ↓
 Cursor
   ↓
 Database

The cursor:

Executes SQL commands
Fetches records
Inserts data
Updates data
Deletes data
Example
import sqlite3

conn = sqlite3.connect("student.db")

cursor = conn.cursor()

Now:

cursor.execute("""
CREATE TABLE IF NOT EXISTS students(
    id INTEGER PRIMARY KEY,
    name TEXT
)
""")

The cursor sends this SQL command to SQLite.

4. What is execute()?

Used to run SQL queries.

cursor.execute("SELECT * FROM students")

or

cursor.execute(
    "INSERT INTO students(name) VALUES(?)",
    ("Prince",)
)
5. What is commit()?

After inserting/updating data:

conn.commit()

This permanently saves changes.

Think of:

execute()
   ↓
Temporary Change
   ↓
commit()
   ↓
Saved Forever

Without commit(), changes may be lost.

6. What is fetchone()?

Gets one row.

cursor.execute("SELECT * FROM students")

row = cursor.fetchone()

print(row)

Output:

(1, 'Prince')
7. What is fetchall()?

Gets all rows.

cursor.execute("SELECT * FROM students")

rows = cursor.fetchall()

print(rows)

Output:

[
 (1,'Prince'),
 (2,'Rahul'),
 (3,'Amit')
]
8. What is close()?

Close the connection when finished.

conn.close()

Think of it as:

Open Database
     ↓
Work
     ↓
Close Database

```python
cursur.execute('''
INSERT INTO employees(name, age, department)
VALUES ('Prince', 20, 'Datascience')
''')

cursur.execute('''
INSERT INTO employees(name, age, department)
VALUES ('Princeee', 20, 'Datascience')
''')

cursur.execute('''
INSERT INTO employees(name, age, department)
VALUES ('Soni', 20, 'Datascience')
''')

conn.commit()
```

```python
cursur.execute('''  
Select * from employees

''')
rows=cursur.fetchall()
for row in rows:
    print(row)
 
 
```

```python
import sqlite3

connection = sqlite3.connect('sales_data.db')
cursor = connection.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS sales (
    id INTEGER PRIMARY KEY,
    date TEXT NOT NULL,
    product TEXT NOT NULL,
    sales INTEGER,
    region TEXT
)
''')

connection.commit()
```

```python
sales_data = [
    ('2023-01-01', 'Product1', 100, 'North'),
    ('2023-01-02', 'Product2', 200, 'South'),
    ('2023-01-03', 'Product1', 150, 'East'),
    ('2023-01-04', 'Product3', 250, 'West'),
    ('2023-01-05', 'Product2', 300, 'North')
]
cursor.executemany('''
INSERT INTO sales(date,product,sales,region)
VALUES(?,?,?,?)               
''',sales_data)
connection.commit()
```

```python
cursor.execute('SELECT * from sales')
rows=cursor.fetchall()
for row in rows:
    print(row)
```

cursor.execute()      # Execute one SQL query
cursor.executemany()  # Execute multiple inserts
cursor.fetchall()     # Fetch all rows
cursor.fetchone()     # Fetch one row
cursor.fetchmany()    # Fetch specified number of rows

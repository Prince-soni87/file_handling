Exception handling is event that distrup the normal flow of a program , they occure  when  an error  is  encountered  diung program  execution. Common  exceptions include 


```python
try:
    a=b
except NameError as e:
    print(f"An error occurred: {e}")    
```

```python
try:
    result=10/0
except ZeroDivisionError as e:
    print(f"An error occurred: {e}")    
```

```python
try:
    result=10/2
    a=b
except ZeroDivisionError as e:
    print(f"An error occurred: {e}")    
except Exception as e:## parent exception class always written in last
    print(f"An error occurred: {e}")    
```

```python
try:
    number=int(input("Enter the NUmber:"))
    result=10/number
except ValueError as e:
    print(f"An error occurred: {e}")
except ZeroDivisionError as e:  
    print(f"An error occurred: {e}")   
except Exception as e:## parent exception class always written in last
    print(f"An error occurred: {e}")     
else:
    print(f"The result is: {result}")## when there is no error then else block will execute    
```

```python
## try,except block with finally
try:
    number=int(input("Enter the NUmber:"))
    result=10/number
except ValueError as e:
    print(f"An error occurred: {e}")
except ZeroDivisionError as e:  
    print(f"An error occurred: {e}")   
except Exception as e:## parent exception class always written in last
    print(f"An error occurred: {e}")     
else:
    print(f"The result is: {result}")## when there is no error then else block will execute    
finally:
    print("This block will always execute")## it allays execute whether there is an error or not
```

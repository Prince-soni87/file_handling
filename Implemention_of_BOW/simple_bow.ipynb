````python
import pandas as pd
messages=pd.read_csv('SMSSpamCollection.txt',
                    sep='\t',names=["label","message"])
````

````python
## Data Cleaning And Preprocessing
import re
import nltk
nltk.download('stopwords')
````

nltk

 │      └── PorterStemmer
 ├── tokenize

````python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
````

````python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
````

# Interview Questions

### Q1. What is CountVectorizer?

**Answer:**
CountVectorizer converts text into numerical vectors by counting the occurrence of words in a document. It is used to create a Bag of Words representation.

---

### Q2. What is `max_features`?

**Answer:**
It limits the vocabulary to the most frequent words. This reduces dimensionality, memory usage, and training time.

---

### Q3. What is `binary=True`?

**Answer:**
Instead of storing the number of occurrences of a word, it stores only whether the word is present (`1`) or absent (`0`).

## Memory Trick

```
CountVectorizer.fit()
        ↓
Learns Words (Vocabulary)

CountVectorizer.transform()
        ↓
Converts Text → Numbers

model.fit()
        ↓
Learns Numbers → Prediction
```

### One sentence to remember forever:

* **`CountVectorizer.fit_transform()` = Train the text preprocessor (learn vocabulary).**
* **`model.fit()` = Train the machine learning model (learn patterns for prediction).**

This distinction is one of the most frequently asked concepts in NLP and Scikit-learn interviews.

# Step 1: Problem with Normal Bag of Words (Unigram)

Suppose we have two sentences:

```text
S1 = The food is good
S2 = The food is not good
```

After preprocessing (remove stopwords like "the", "is"):

```text
S1 = food good
S2 = food not good
```

Vocabulary becomes:

```text
food
not
good
```

BoW matrix:

| Sentence | food | not | good |
| -------- | ---- | --- | ---- |
| S1       | 1    | 0   | 1    |
| S2       | 1    | 1   | 1    |

Notice something?

Both sentences look **almost the same**.

The only difference is:

```text
not
```

But in reality,

```text
food good
```

means **positive** 😊

while

```text
food not good
```

means **negative** 😞

The normal Bag of Words cannot understand that **"not good"** is a phrase.

---

# Step 2: Solution → N-Grams

Instead of looking at **one word**, look at **multiple consecutive words**.

---

## Unigram (1 word)

Sentence

```text

```

Tokens:

```text

not
good
```

These are called **Unigrams**.

---

## Bigram (2 words)

Now combine every two consecutive words.

Sentence

```text

```

Becomes

```text

```

These are called **Bigrams**.

Notice:

```text

```

is now treated as a **single feature**.

Now the model can understand

```text
not good
```

has a different meaning than

```text
good
```

# Why not always use Trigram or 4-Gram?

Because the vocabulary grows very quickly.

Example:

Suppose you have

```text
10,000 unique words
```

Unigrams:

```
10,000 features
```

Unigrams + Bigrams:

```
Hundreds of thousands of features
```

Unigrams + Bigrams + Trigrams:

```
Millions of features
```

This makes training slower and uses much more memory.

---

# When to use which?

### Unigram `(1,1)`

✅ Small datasets

✅ Simple text classification

---

### Unigram + Bigram `(1,2)` ⭐

Most commonly used.

Captures phrases like

```text
not good
very happy
machine learning
```

---

### Unigram + Bigram + Trigram `(1,3)`

Useful when phrases of three words matter, such as:

```text
New York City
United States Government
I love Python
```

# Interview Questions

### Q1. What is an N-Gram?

**Answer:**
An N-Gram is a sequence of **N consecutive words** extracted from a sentence.

---

### Q2. Why do we use Bigrams?

**Answer:**
Bigrams preserve local context by treating two consecutive words (such as "not good") as a single feature, which helps capture meaning that unigrams may miss.

**Answer:**
It includes both unigrams (single words) and bigrams (pairs of consecutive words).


## Rule to remember

* **Unigram** → One word (less context, fewer features)
* **Bigram** → Two words (captures short phrases like *not good*)
* **Trigram** → Three words (captures longer expressions)
* Increasing **N** gives **more context** but also **more features**, which increases memory usage and training time.

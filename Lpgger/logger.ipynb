Logging is a crucial aspect of any application, providing a way to track events, errors, and operational information. Python's built-in logging module offers a flexible framework for emitting log messages from Python programs. In this lesson, we will cover the basics of logging, including how to configure logging, log levels, and best practices for using logging in Python applications.

```python
import logging

## Configure the basic logging settings
logging.basicConfig(level=logging.DEBUG)

## log messages with different severity levels
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```

Python's logging module has several log levels indicating the severity of events. The default levels are:

DEBUG: Detailed information, typically of interest only when diagnosing problems.
INFO: Confirmation that things are working as expected.
WARNING: An indication that something unexpected happened or indicative of some problem in the near future (e.g., "disk space low"). The software is still working as expected.
ERROR: Due to a more serious problem, the software has not been able to perform some function.
CRITICAL: A very serious error, indicating that the program itself may be unable to continue running.


| Log Level | Purpose                                          | Example                         |
| --------- | ------------------------------------------------ | ------------------------------- |
| DEBUG     | Detailed debugging information                   | Variable values, loop execution |
| INFO      | Normal program execution messages                | "Application started"           |
| WARNING   | Something unusual happened but program continues | "Disk space low"                |
| ERROR     | A function or operation failed                   | "Database connection failed"    |
| CRITICAL  | Severe error, program may crash                  | "Server shutdown"               |


```python
import logging

logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)

logging.debug("Debug message")
logging.info("Info message")
logging.warning("Warning message")
logging.error("Error message")
logging.critical("Critical message")
```

What is logging.getLogger()?

getLogger() is a function in Python's logging module that creates or returns a logger object.

import logging

logger = logging.getLogger("ArithmeticApp")

Here:

"ArithmeticApp" is the name of the logger.
logger is an object used to write log messages.
Why use getLogger()?

Instead of writing:

logging.debug("Adding numbers")
logging.error("Something went wrong")

you can create a dedicated logger:

logger = logging.getLogger("ArithmeticApp")

logger.debug("Adding numbers")
logger.error("Something went wrong")

Now every log message will contain the logger name:

2026-06-19 12:00:00 - ArithmeticApp - DEBUG - Adding numbers
In your code
logger = logging.getLogger("ArithmeticApp")

creates a logger named ArithmeticApp.

Then:

logger.debug(f"Subtracting {a} - {b} = {result}")

produces:

2026-06-19 12:00:00 - ArithmeticApp - DEBUG - Subtracting 10 - 5 = 5
What happens if I call getLogger() twice?
logger1 = logging.getLogger("ArithmeticApp")
logger2 = logging.getLogger("ArithmeticApp")

Both variables refer to the same logger object.

print(logger1 is logger2)

Output:

True

__name__ vs Custom Name

Custom name:

logger = logging.getLogger("ArithmeticApp")

Automatic module name:

logger = logging.getLogger(__name__)

In ML projects and internships, you'll usually see:

logger = logging.getLogger(__name__)

because it automatically uses the file name (data_ingestion, model_trainer, etc.) as the logger name.

Simple Definition

logging.getLogger(name) creates or retrieves a named logger object that is used to generate log messages for a specific application, module, or file.

```python
import logging

logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    handlers=[
        logging.FileHandler("app1.log"),
        logging.StreamHandler()
    ]
)

logger = logging.getLogger("ArithmeticApp")

def add(a, b):
    result = a + b
    logger.debug(f"Adding {a}+{b}", result)
    return result

def subtract(a, b):
    result = a - b
    logger.debug(f"Subtracting {a} - {b} = {result}")
    return result

def multiply(a, b):
    result = a * b
    logger.debug(f"Multiplying {a} * {b} = {result}")
    return result

def divide(a, b):
    try:
        result = a / b
        logger.debug(f"Dividing {a} / {b} = {result}")
        return result
    except ZeroDivisionError:
        logger.error("Division by zero error")
        return None
```

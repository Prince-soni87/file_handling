# What is Lemmatization?

**Definition:**

> Lemmatization reduces a word to its dictionary root form (called a lemma).

Unlike stemming:

```text id="0h4eeq"
Stemming = Cuts words

Lemmatization = Understands words
```

# Specify Part of Speech (POS)

### Verb

```python id="4qws48"
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("running", pos='v'))
print(lemmatizer.lemmatize("playing", pos='v'))
```

Output:

```text id="4byhy9"
run
play
```

Now it works correctly.

# More Examples

```python id="jlwmc6"
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = [
    "running",
    "playing",
    "better",
    "studies"
]

for word in words:
    print(word, "->", lemmatizer.lemmatize(word, pos='v'))
```

Output:

```text id="r4utyn"
running -> run
playing -> play
better -> better
studies -> study
```

# Interview Question

### What is Lemmatization?

**Answer:**

Lemmatization is the process of reducing a word to its dictionary root form (lemma) while preserving its meaning.

---

### Difference Between Stemming and Lemmatization?

| Stemming                 | Lemmatization                  |
| ------------------------ | ------------------------------ |
| Cuts words using rules   | Uses vocabulary and dictionary |
| Faster                   | Slower                         |
| May create invalid words | Produces valid words           |
| Example: studies → studi | Example: studies → study       |

---

### Why is Lemmatization Better?

Because it returns meaningful dictionary words.

Example:

```text id="q8zjpl"
Stemming:
history → histori

Lemmatization:
history → history
```

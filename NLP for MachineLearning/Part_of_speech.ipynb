# What is Part of Speech (POS)?

In English grammar, every word has a role.

Example:

```text id="ys8vcc"
Prince plays cricket.
```

| Word    | Role |
| ------- | ---- |
| Prince  | Noun |
| plays   | Verb |
| cricket | Noun |

These roles are called:

```text id="mujkw1"
Part of Speech (POS)
```

---

# Common POS Tags

| POS | Meaning     | Example           |
| --- | ----------- | ----------------- |
| NN  | Noun        | boy, car, cricket |
| VB  | Verb        | play, eat, run    |
| JJ  | Adjective   | beautiful, smart  |
| RB  | Adverb      | quickly, slowly   |
| PRP | Pronoun     | I, he, she        |
| DT  | Determiner  | the, a, an        |
| IN  | Preposition | in, on, at        |

---

# Example

Sentence:

```text id="y3m7ic"
The boy is playing cricket.
```

POS Tags:

```text id="6yup1x"
The      → DT
boy      → NN
is       → VBZ
playing  → VBG
cricket  → NN
```

### Code

```python id="igv06q"
from nltk.tokenize import word_tokenize
from nltk import pos_tag

sentence = "The boy is playing cricket"

words = word_tokenize(sentence)

print(pos_tag(words))
```

Output:

```python id="v1u7g2"
[
 ('The', 'DT'),
 ('boy', 'NN'),
 ('is', 'VBZ'),
 ('playing', 'VBG'),
 ('cricket', 'NN')
]
```

# Why is POS Tagging Important?

Suppose the word is:

```text id="0nwzjn"
book
```

Sentence 1:

```text id="3o6m3o"
I read a book.
```

Here:

```text id="z9z6oi"
book = Noun
```

---

Sentence 2:

```text id="c0k95i"
Book a ticket.
```

Here:

```text id="ixz5ko"
book = Verb
```

Same word, different meaning.

POS Tagging helps the computer understand this.

# Real-Life Example

Sentence:

```text id="3b5hlv"
I am learning NLP.
```

After tokenization:

```python id="pypghm"
['I', 'am', 'learning', 'NLP']
```

After POS tagging:

```python id="m7k39t"
[
 ('I', 'PRP'),
 ('am', 'VBP'),
 ('learning', 'VBG'),
 ('NLP', 'NN')
]
```

Now the computer knows:

```text id="yudjdl"
I → Pronoun
am → Verb
learning → Verb
NLP → Noun
```

# Why Needed in Lemmatization?

Remember:

```python id="h1yrkq"
lemmatizer.lemmatize("running")
```

Output:

```text id="lqwh4t"
running
```

Not:

```text id="2e3zcc"
run
```

Because WordNet assumes:

```text id="m6i2lk"
running = noun
```

---

If we specify:

```python id="vmojhr"
lemmatizer.lemmatize("running", pos='v')
```

Output:

```text id="suyc83"
run
```

POS tagging helps lemmatization choose the correct root word.

# Interview Questions

### What is POS Tagging?

**Answer:**

POS Tagging is the process of assigning grammatical categories (noun, verb, adjective, etc.) to each word in a sentence.

---

### Why is POS Tagging important?

**Answer:**

1. Helps understand sentence structure.
2. Improves lemmatization.
3. Used in Named Entity Recognition (NER).
4. Helps machine translation.
5. Helps sentiment analysis and text classification.

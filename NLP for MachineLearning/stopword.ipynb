# What are Stopwords?

Stopwords are **common words** that occur very frequently in a language but usually do not add much meaning to the text.

Examples:

```text id="p7a4n9"
is
am
are
the
a
an
in
on
at
of
to
and
for
```

# Why Remove Stopwords?

Suppose we have 1000 documents.

Words like:

```text id="7w59rn"
the
is
am
are
```

appear everywhere.

These words don't help much in understanding the topic.

Example:

```text id="jj88gt"
The movie is good.
The food is good.
The course is good.
```

Words:

```text id="l71xjv"
the
is
```

appear everywhere.

Removing them helps the model focus on:

```text id="8nkvci"
movie
food
course
good
```

# Remove Stopwords

```python id="hr5lh4"
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

sentence = "I am learning NLP from NLTK"

words = word_tokenize(sentence)

filtered_words = [
    word for word in words
    if word.lower() not in stopwords.words('english')
]

print(filtered_words)
```

Output:

```python id="f8ibz6"
['learning', 'NLP', 'NLTK']
```

# Interview Question

### What are Stopwords?

**Answer:**

Stopwords are commonly occurring words in a language that carry little semantic meaning and are often removed during text preprocessing.

---

### Why Remove Stopwords?

**Answer:**

* Reduce text size
* Improve processing speed
* Reduce noise
* Focus on meaningful words

---

# Important Interview Trick Question

### Should we always remove stopwords?

**Answer: No.**

Example:

```text id="o7o70h"
I do not like this movie.
```

If you remove stopwords:

```text id="fbbnl6"
like movie
```

The word:

```text id="uxs61j"
not
```

is extremely important because it changes the meaning.

So in:

* Sentiment Analysis
* Chatbots
* Question Answering

we often keep some stopwords.

````python

from nltk.tokenize import word_tokenize
for i in range(len(sentence)):
    words=word_tokenize(sentence[i])
    words=[stemm.stem(word) for word in words if word not in stopwords.words('english')]
    sentence[i]=' '.join(words)


````

````python
from nltk.stem import SnowballStemmer
snowballstemmer=SnowballStemmer('english')
````

````python
## Apply Stopwords And Filter And then Apply Snowball Stemming

for i in range(len(sentence)):
    words=nltk.word_tokenize(sentence[i])
    words=[snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentence[i]=' '.join(words)# converting all the list of words into sentences
````

````python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
````

````python
## Apply Stopwords And Filter And then Apply Snowball Stemming

for i in range(len(sentence)):
    #sentences[i]=sentences[i].lower()
    words=nltk.word_tokenize(sentence[i])
    words=[lemmatizer.lemmatize(word.lower(),pos='v') for word in words if word not in set(stopwords.words('english'))]
    sentence[i]=' '.join(words)# converting all the list of words into sentences
````

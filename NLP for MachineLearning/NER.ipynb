# What is Named Entity Recognition (NER)?

NER is the process of identifying and classifying important entities in text.

Examples of entities:

| Entity Type | Example |
|------------|----------|
| Person | Prince Soni, Elon Musk |
| Organization | OpenAI, Microsoft |
| Location | Delhi, India, New York |
| Date | 15 August 1947 |
| Time | 10:30 AM |
| Money | ₹5000, $100 |
| Percentage | 85% |

---

# Example

Sentence:

```text id="m6gbjk"
Prince Soni studies at GLA University in Mathura.
```

NER Output:

```text id="l4h9sw"
Prince Soni      → PERSON
GLA University   → ORGANIZATION
Mathura          → LOCATION
```

---

# Difference Between POS Tagging and NER

### POS Tagging

Sentence:

```text id="w6myv4"
Prince studies at GLA University.
```

Output:

```text id="lj5zv5"
Prince   → Noun
studies  → Verb
GLA      → Proper Noun
University → Proper Noun
```

POS only tells grammar.

---

### NER

Output:

```text id="7k6yjp"
Prince → PERSON
GLA University → ORGANIZATION
```

NER tells the real-world meaning.

### Example Code

```python id="0k36rn"
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk

sentence = "Prince Soni studies at GLA University in Mathura"

words = word_tokenize(sentence)

pos_tags = pos_tag(words)

named_entities = ne_chunk(pos_tags)

print(named_entities)
```

Output (tree-like):

```text id="h53v96"
(PERSON Prince/NNP Soni/NNP)
(ORGANIZATION GLA/NNP University/NNP)
(GPE Mathura/NNP)
```

# Interview Questions

### What is NER?

**Answer:**

Named Entity Recognition is an NLP technique used to identify and classify named entities such as people, organizations, locations, dates, money, and other real-world objects in text.

---

### Why is NER used?

**Answer:**

- Information extraction
- Search engines
- Chatbots
- Resume parsing
- News analysis
- Medical NLP
- Financial document analysis

---

### Difference Between POS and NER?

| POS Tagging | NER |
|------------|-----|
| Finds grammatical role | Finds real-world entity |
| Noun, Verb, Adjective | Person, Organization, Location |
| Syntax focused | Meaning focused |

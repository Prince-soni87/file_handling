Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma. Stemming is important in natural language understanding (NLU) and natural language processing (NLP).

````python
## Classification Problem
## Comments of product is a positive review or negative review
## Reviews----> eating, eat,eaten [going,gone,goes]--->go

words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
````

````python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()

````

````python
for word in words:
    print(word+"---->"+stemmer.stem(word))
````

````python
stemmer.stem('congratulations')
````

````python
stemmer.stem("sitting")
````

NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression. Let us see an example

Excellent! You're now learning **RegexpStemmer**, which is a special stemmer in NLTK.

# What is RegexpStemmer?

**RegexpStemmer** uses a **Regular Expression (Regex)** pattern to remove prefixes or suffixes from words.

Unlike PorterStemmer:

```text id="3kj3n2"
PorterStemmer → Uses predefined rules

RegexpStemmer → Uses YOUR regex pattern
```

Think of it as:

```text id="e1csmu"
You tell the stemmer what to remove.
```

---

# Syntax

```python id="6xb77u"
from nltk.stem import RegexpStemmer

stemmer = RegexpStemmer(pattern)
```

---

# Example 1: Remove "ing"

```python id="0v5z67"
from nltk.stem import RegexpStemmer

stemmer = RegexpStemmer('ing$')

print(stemmer.stem('playing'))
print(stemmer.stem('running'))
print(stemmer.stem('working'))
```

### Output

```text id="65lg1l"
play
runn
work
```

---

# Understanding the Regex

Pattern:

```python id="hmtg1w"
'ing$'
```

Meaning:

```text id="59jvhh"
ing  → letters to match
$    → must be at the end of the word
```

Examples:

```text id="3oz0o5"
playing  ✓
running  ✓
working  ✓
```

---

# Example 2: Remove "ing", "ed", "s"

```python id="4vhzy7"
from nltk.stem import RegexpStemmer

stemmer = RegexpStemmer('ing$|ed$|s$')

words = ['playing', 'played', 'plays']

for word in words:
    print(stemmer.stem(word))
```

### Output

```text id="8tgz0w"

play
play
```

# Why Snowball Stemmer?

Earlier we learned:

```python
from nltk.stem import PorterStemmer
```

Porter Stemmer works well, but sometimes produces strange stems.

Example:

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

print(stemmer.stem("fairly"))
print(stemmer.stem("sportingly"))
```

Output:

```text
fairli
sportingli
```

Notice:

```text
fairli ❌
sportingli ❌
```

These are not proper words.

To improve this, Snowball Stemmer was introduced.

---

# What is Snowball Stemmer?

**Definition:**

> Snowball Stemmer is an improved version of Porter Stemmer that provides more accurate stemming and supports multiple languages.

Also called:

```text
Porter2 Stemmer
```

because it is the improved version of Porter.

---

# Example

```python
from nltk.stem import SnowballStemmer

snowball = SnowballStemmer("english")

print(snowball.stem("running"))
print(snowball.stem("playing"))
print(snowball.stem("fairly"))
```

Output:

```text
run
play
fair
```

---

# Compare Porter vs Snowball

```python
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer

porter = PorterStemmer()
snowball = SnowballStemmer("english")

word = "fairly"

print("Porter :", porter.stem(word))
print("Snowball:", snowball.stem(word))
```

Output:

```text
Porter : fairli
Snowball: fair
```

Snowball gives a cleaner result.

---

```text

   │
   ├── Porter Stemmer
```

# Important Note

Even Snowball Stemmer can produce non-dictionary words.

Example:

```python
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer("english")

print(stemmer.stem("easily"))
```

Output:

```text
easili
```

Notice:

```text
easili
```

is not a real English word.

That's because:

```text
Stemming = Rule-Based Cutting
```

while:

```text
Lemmatization = Dictionary-Based Understanding
```
